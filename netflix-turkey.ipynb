```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
Turkey = df[df.country=='Turkey']
```

```python
Turkey.head()
```

```python
sns.countplot(x = 'type', data = Turkey)
```

```python
plt.figure(figsize=(15,7))
ratings_in = Turkey.groupby(['rating'])['show_id'].count().reset_index(name = 'count').sort_values(by='count',ascending=False)
sns.pointplot(x= 'rating', y= 'count', data=ratings_in)
```

```python
plt.figure(figsize=(25,7))
top_category = Turkey.groupby(['listed_in'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False) 
sns.pointplot(x='listed_in',y='count',data=top_category.head(5),color='green')
```

```python
plt.figure(figsize=(25,7))
top_director = Turkey.groupby(['director'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
sns.pointplot(x='director',y='count',data=top_director.head(10))
```

```python
plt.figure(figsize=(15,7))
Turkey_recent5 = Turkey[Turkey.release_year >= 2011]
g = sns.catplot(data= Turkey_recent5, kind='count', x = 'release_year', hue='type')
```

```python
movies = Turkey[Turkey.type == 'Movie']
TVShows = Turkey[Turkey.type == 'TV Show']
```

```python
plt.figure(figsize=(15,7))
_10yrs = movies[movies.release_year >= 2011]
sns.countplot(x= 'release_year', data=_10yrs)
```

```python
plt.figure(figsize=(25,7))
director = movies.groupby(['director'])['show_id'].count().reset_index(name = 'count').sort_values(by = 'count', ascending = False)
sns.lineplot(x = 'director', y= 'count', data=director.head(10))
```

```python
plt.figure(figsize=(15,7))
ratings = movies.groupby(['rating'])['show_id'].count().reset_index(name = 'count').sort_values(by = 'count', ascending = False)
sns.pointplot(x = 'rating', y= 'count', data=ratings)
```

```python
plt.figure(figsize=(25,7))
listed = movies.groupby(['listed_in'])['show_id'].count().reset_index(name = 'count').sort_values(by ='count', ascending = False)
sns.pointplot(x = 'listed_in', y= 'count', data=listed.head(5), color = 'green')
```

```python
plt.figure(figsize=(10,7))
_10yrsTV = TVShows[TVShows.release_year >= 2011]
sns.countplot(x = 'release_year', data = _10yrsTV)
```

```python
plt.figure(figsize=(35,7))
director_TV = TVShows.groupby(['director'])['show_id'].count().reset_index(name = 'count').sort_values(by = 'count', ascending = False)
sns.lineplot(x = 'director', y = 'count', data= director_TV.head(10))
```
